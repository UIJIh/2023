# ML

## **지도학습 (Supervised Learning):**
1. **분류 (Classification):**
   - 결정 트리 (Decision Tree): 데이터를 여러 개의 분기로 나누어 클래스를 분류하는 모델입니다. 예측을 위한 질문들을 계속하여 더 잘 나눠나가는 방식으로 작동합니다.
   - 앙상블 학습 (Ensemble Learning):
     - 랜덤 포레스트 (Random Forest): 다수의 결정 트리를 조합한 모델로, 각 트리의 예측 결과를 투표하여 최종 예측을 결정합니다.
     - 그래디언트 부스팅 (Gradient Boosting): 여러 개의 결정 트리를 순차적으로 학습시켜 오차를 보정하는 방식으로 예측을 개선합니다.
     - XGBoost, LightGBM, CatBoost 등: 그래디언트 부스팅 기반의 알고리즘으로 속도와 성능을 향상시킨 모델들입니다.

2. **회귀 (Regression):**
   - 선형 회귀 (Linear Regression): 입력 데이터와 가중치의 선형 결합으로 예측 값을 계산하는 모델입니다.
   - 릿지 회귀 (Ridge Regression) 및 라쏘 회귀 (Lasso Regression): 선형 회귀의 변형으로, L2 정규화와 L1 정규화를 사용하여 모델을 규제합니다.

## **비지도학습 (Unsupervised Learning):**
1. **군집화 (Clustering):**
   - K-평균 군집화 (K-Means Clustering): 데이터를 K개의 클러스터로 그룹화하는 모델입니다. 각 데이터 포인트를 가장 가까운 클러스터 중심에 할당합니다.
   - DBSCAN (Density-Based Spatial Clustering of Applications with Noise): 데이터의 밀도에 따라 클러스터를 식별하는 모델로, 노이즈 데이터를 식별할 수 있습니다.

2. **차원 축소 (Dimensionality Reduction):**
   - 주성분 분석 (Principal Component Analysis, PCA): 데이터의 주요 정보를 유지하면서 차원을 축소하는 모델로, 변수들 간의 상관관계를 이용하여 새로운 변수를 생성합니다.

### **평가지표 (Evaluation Metrics):**
- **분류 평가지표:**
   - 정확도 (Accuracy): 전체 예측 중 정확하게 예측한 비율입니다.
   - 정밀도 (Precision): 양성 예측 중 실제 양성인 비율입니다.
   - 재현율 (Recall): 실제 양성 중 양성으로 예측한 비율입니다.
   - F1 스코어 (F1-Score): 정밀도와 재현율의 조화 평균입니다.

- **회귀 평가지표:**
   - 평균 제곱근 오차 (Root Mean Squared Error, RMSE): 실제 값과 예측 값 사이의 차이를 평균화한 값입니다.
   - 평균 절대 오차 (Mean Absolute Error, MAE): 실제 값과 예측 값 사이의 절대값 차이의 평균입니다.

````python
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_squared_error
import numpy as np

# 랜덤 포레스트 모델 생성
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train.reshape(-1, 1), y_train)

# XGBoost 모델 생성
xgb_model = XGBRegressor(n_estimators=100, random_state=42)
xgb_model.fit(X_train.reshape(-1, 1), y_train)

# LightGBM 모델 생성
lgbm_model = LGBMRegressor(n_estimators=100, random_state=42)
lgbm_model.fit(X_train.reshape(-1, 1), y_train)

# 모델 예측
rf_predictions = rf_model.predict(X_train.reshape(-1, 1))
xgb_predictions = xgb_model.predict(X_train.reshape(-1, 1))
lgbm_predictions = lgbm_model.predict(X_train.reshape(-1, 1))

# RMSE 평가 지표 계산
rf_rmse = np.sqrt(mean_squared_error(y_train, rf_predictions))
xgb_rmse = np.sqrt(mean_squared_error(y_train, xgb_predictions))
lgbm_rmse = np.sqrt(mean_squared_error(y_train, lgbm_predictions))

print("Random Forest RMSE:", rf_rmse)
print("XGBoost RMSE:", xgb_rmse)
print("LightGBM RMSE:", lgbm_rmse)
````

## 차수 추정 (시계열 모형에서)

1. `from statsmodels.tsa.stattools import pacf`: `statsmodels` 패키지에서 시계열 데이터의 PACF를 계산하는 함수인 `pacf`를 불러옵니다.

2. `pacf_values = pacf(df['cnt'], nlags=20, method='ols')`: `df['cnt']` 열에 대한 20개의 시차(lags)에 대한 PACF 값을 계산하여 `pacf_values` 변수에 저장합니다. `method='ols'`는 최소 자승법(Ordinary Least Squares)을 사용하여 계산하겠다는 의미입니다.

5. `plot_pacf(df['cnt'], lags=20, method='ols', title='pa').show()`: `df['cnt']` 열에 대한 PACF를 시각화합니다. `lags=20`은 최대 20개의 시차까지 시각화하겠다는 의미이며, `method='ols'`는 PACF를 계산할 때 최소 자승법을 사용하겠다는 의미입니다. 마지막으로 `.show()`를 호출하여 시각화된 그래프를 표시합니다.

## Scaling 전처리
- 정규분포에 가깝도록
- faster

1. StandardScaler

평균=0, 표준편차=1로 조정해서 모든 특성이 같은 크기를 갖게 한다.

 

#### 2. MinMaxScaler

최대값=1, 최소값=0으로 조정, 아웃라이어(이상치)에 취약

 

3. RobustScaler

중앙값=0, IQR(1분위~3분위값)=1로 조정, 아웃라이어 영향을 최소화하며 정규분포보다 더 넓게 분포

 

4. MaxAbsScaler

0을 기준으로 절대값이 가장 큰 수가 1 또는 -1이 되도록 조정, 양수 데이터로만 구성된 데이터셋에서는 아웃라이어에 민감

 

데이터 스케일링을 하는 이유 = 아웃라이어의 영향을 최소화하는 것이기 때문에 보통은 이상치 영향을 가장 적게 받는 StandardScaler 혹은 RobustScaler를 주로 사용

#### fit_transform()
- fit()과 transform()을 한번에 처리할 수 있게 하는 메서드
- 테스트 데이터에는 fit_transform()메서드를 쓰면 안됨
- fit()은 데이터를 학습시키는 메서드 / transform()은 실제로 학습시킨 것을 적용하는 메서드

````python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0,1))
train_scaled = sc.fit_transform(train)
````

````python
train_scaled
````

````python
import numpy as np

# 자기 자신 과거를 독립변수로 사용하는 시계열

X_train = []
y_train = []

for i in range(1, 17000):
    X_train.append(train_scaled[i-1:i, 0])   # row 는 1시간전~
    y_train.append(train_scaled[i, 0])       # col 은 그대로

X_train, y_train = np.array(X_train), np.array(y_train)
````

````python
X_train.shape
````

시계열 데이터를 LSTM 네트워크에 입력하기 위해 데이터의 형태를 변환하는 과정을 수행

1. `X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))`
: `X_train` 데이터의 형태를 변환하여 LSTM 네트워크에 적합한 형태로 만듭니다. LSTM 모델은 3D 형태의 입력을 사용하며, `(samples, timesteps, features)` 형태를 가집니다. 여기서 각 차원은 다음과 같이 해석됩니다:
   - `samples`: 입력 데이터의 총 샘플 수
   - `timesteps`: 시계열 데이터의 각 시간 단계
   - `features`: 각 시간 단계에서 사용되는 특성의 수

   `np.reshape()` 함수를 사용하여 `X_train`의 형태를 변환합니다. 이때, `X_train.shape[0]`은 데이터의 총 샘플 수를 나타내고, `X_train.shape[1]`은 각 샘플에서 사용된 시간 단계의 수를 나타냅니다. 마지막의 `1`은 각 시간 단계에서 사용되는 특성의 수를 나타냅니다. 이 코드를 통해 `X_train`은 LSTM 모델에 입력될 수 있는 형태로 변환됩니다.

이렇게 형태를 변환하는 이유는 LSTM 모델이 시퀀스 데이터를 처리하기 위해 과거 시점의 정보를 활용하기 때문입니다. `X_train`은 시간 단계별로 구성된 데이터이며, 이를 LSTM 모델에 적합한 형태로 재구성하는 것이 목적입니다.

````python
# 3차원으로
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1],1))
X_train.shape
````

Keras 라이브러리를 사용하여 순환 신경망 (RNN) 모델을 생성하는 과정을 나타냅니다.

1. `from keras.models import Sequential`: Keras에서 모델을 생성하는 데 사용되는 `Sequential` 클래스를 불러옵니다. `Sequential` 클래스는 순차적인 레이어를 쉽게 추가할 수 있게 해줍니다.

2. `from keras.layers import Dense, Activation, SimpleRNN`: Keras에서 사용되는 레이어들을 불러옵니다.
   - `Dense` 레이어는 완전 연결층 (fully connected layer)을 생성하는 데 사용됩니다.
   - `Activation`은 활성화 함수를 설정하는 데 사용됩니다.
   - `SimpleRNN`은 단순한 순환 신경망 레이어를 생성하는 데 사용됩니다.

**층쌓기**
이제 위의 import 구문을 사용하여 모델을 생성하고 레이어를 추가하는 코드를 작성할 수 있습니다. 아래는 기본적인 순서입니다:

```python
model = Sequential()  # Sequential 모델 생성

# Simple RNN 레이어 추가
model.add(SimpleRNN(units=hidden_units, activation='relu', input_shape=(timesteps, features)))

# 출력층 추가
model.add(Dense(output_units))  # 출력층
model.add(Activation('linear'))  # 활성화 함수 설정
```

위 코드에서 `hidden_units`는 Simple RNN의 유닛 수를, `timesteps`는 시간 단계의 수를, `features`는 입력 데이터의 특성 수를 나타냅니다. `output_units`는 출력층의 유닛 수를 나타냅니다.

마지막으로, 모델을 컴파일하고 학습할 수 있습니다. 학습 관련 코드는 이후에 추가해야 할 것입니다.

   - `batch_size=1`: 배치 크기를 1로 설정합니다. 한 번의 역전파마다 한 샘플씩 업데이트됩니다.
   - `epochs=2`: 전체 데이터를 2번 반복해서 학습합니다.

배치 크기(Batch size)는 신경망 모델을 한 번의 학습 업데이트에 사용되는 데이터 샘플의 수를 나타냅니다. 간단히 말해서, 한 번의 모델 업데이트를 위해 몇 개의 데이터 샘플을 동시에 사용할 것인지를 결정하는 파라미터입니다.

배치 크기가 큰 경우:
- 학습이 빠르게 진행될 수 있습니다.
- 하드웨어 자원을 효율적으로 활용할 수 있습니다.
- 그러나 메모리 사용량이 크게 증가할 수 있습니다.

배치 크기가 작은 경우:
- 학습이 더 안정적으로 이루어질 수 있습니다.
- 메모리 사용량이 줄어들어 더 작은 하드웨어에서도 학습이 가능합니다.
- 학습 속도는 상대적으로 느릴 수 있습니다.

일반적으로 배치 크기는 데이터셋의 크기에 따라 결정되며, 너무 크지도 작지도 않은 적당한 값을 선택하는 것이 중요합니다. 보통 32, 64, 128 등의 배치 크기가 많이 사용되며, 경험적으로 좋은 성능을 보이는 값을 선택합니다.

좋은 배치 크기를 선택하는 과정은 모델과 데이터에 따라 다를 수 있으며, 실험을 통해 최적의 값을 찾아내는 것이 좋습니다.

````python
rnn.fit(X_train, y_train, batch_size=1, epochs=2)
````

- `16999/16999`: 현재까지 처리한 배치의 수와 전체 배치 수입니다. 첫 번째 값은 현재까지 처리한 배치의 수이고, 두 번째 값은 전체 배치 수입니다.

- `[==============================]`: 진행 상태를 나타내는 그래프 형태의 표시입니다. 이 그래프는 학습 진행 상태를 시각화한 것이며, `[`와 `]` 사이의 부분이 학습의 진행 정도를 나타냅니다.

- `34s`: 학습 소요 시간을 나타냅니다. 이 경우에는 학습에 34초가 소요되었습니다.

- `2ms/step`: 각 배치 단계에서 소요된 시간을 나타냅니다. 이 경우에는 1 배치를 처리하는 데 약 2밀리초가 소요되었습니다.

- `loss: 0.0068`: 현재까지 계산된 손실 값입니다. 이 값은 평균 제곱 오차(Mean Squared Error) 손실 함수의 결과입니다.
    - 손실 값이 낮을 때: 낮은 손실 값은 모델이 학습 데이터에서 잘 동작하고 있다는 것을 나타냅니다. 하지만 손실 값만 낮다고 해서 모델이 항상 좋은 성능을 내는 것은 아닙니다. 이때, *'과적합'*의 가능성이 있습니다. 학습 데이터에만 지나치게 적합하여 새로운 데이터에 대한 일반화 능력이 떨어질 수 있습니다.
    - 손실 값이 높을 때: 높은 손실 값은 모델이 학습 데이터를 잘 표현하지 못하고 있다는 것을 나타냅니다. 이 경우, *'과소적합(underfitting)'*의 가능성이 있습니다. 모델이 학습 데이터에도 적합하지 않고 새로운 데이터에도 예측 능력이 떨어질 수 있습니다.

- `mse: 0.0068`: 평균 제곱 오차(Mean Squared Error)의 결과를 나타냅니다. 
    - MSE가 작다는 것은 모델이 학습 데이터를 잘 표현하고 있고, 예측이 정확하다는 것을 나타냅니다. 하지만 MSE 값이 너무 작다면, 모델이 학습 데이터에 과적합되었을 가능성도 있습니다. 따라서 항상 학습 데이터뿐만 아니라 검증 데이터나 테스트 데이터에서 모델의 성능을 평가하는 것이 중요합니다.

#### scaling

`sc.transform(test)`는 테스트 데이터인 `test`를 정규화된 형태로 변환하는 작업을 의미합니다. 이 작업은 훈련 데이터를 정규화할 때 사용한 스케일러(예: `MinMaxScaler`나 `StandardScaler`)를 활용하여 테스트 데이터를 동일한 방식으로 정규화하여 모델에 입력하기 위함입니다.

여기서 `sc`는 스케일러 객체를 나타냅니다. 일반적으로 데이터를 정규화할 때는 훈련 데이터에 대해 `fit` 메서드를 사용하여 스케일러를 피팅한 뒤, `transform` 메서드를 사용하여 훈련 데이터를 정규화합니다. 그 후에는 같은 스케일러 객체를 사용하여 테스트 데이터도 정규화합니다.

`inputs = sc.transform(test)`의 결과로는 `test` 데이터가 정규화된 형태로 변환된 `inputs` 데이터가 반환됩니다. 이렇게 정규화된 데이터를 모델에 입력하여 예측을 수행하거나 다른 작업을 진행할 수 있습니다.

이 경우 LSTM 모델을 사용하여 예측을 수행하려면 input_shape를 모델에 맞게 설정해야 함

#### 역 스케일링
- 데이터를 정규화 또는 표준화한 후에 원래의 스케일로 변환하는 작업을 의미합니다. 이 작업은 주로 모델이 예측한 결과를 원래 데이터의 단위나 범위로 변환하기 위해 수행됩니다. 

- 예를 들어, 여러 특성을 가진 데이터를 입력으로 사용하여 모델을 훈련하고 예측을 수행했다고 가정해봅시다. 이때 입력 데이터를 정규화하거나 표준화하여 모델의 성능을 향상시킬 수 있습니다. 하지만 모델이 예측한 결과는 정규화 된 형태이므로 이 결과를 실제 데이터의 단위와 범위로 변환하기 위해서는 역 스케일링이 필요합니다.

- 역 스케일링을 통해 모델의 예측 결과를 실제 값으로 변환하면, 예측한 값의 의미와 실제 데이터 간의 비교가 가능해집니다. 이로써 모델이 얼마나 정확한 예측을 했는지를 보다 명확하게 평가하고 해석할 수 있습니다.

# GRU
- gated recurrent unit
- lstm의 overfiting 문제 보완
- unit의 개수는 줄이고 성능은 개선

````python
from keras.layers import GRU

gru = Sequential()
gru.add(GRU(units=6, activation='relu', input_shape=(1,1)))
gru.add(Dense(units=1, activation='linear'))
````

````python
print(gru.summary())
````

#### Param
RNN < GRU < LSTM

````python
gru.compile(loss='mse', optimizer='adam', metrics=['mse'])
gru.fit(X_train, y_train, batch_size=1, epochs=2)
````

````python
gru = gru.predict(X_test)
gru = sc.inverse_transform(gru)
````

````python
#test = np.array(test)
plt.figure(figsize=(15,5))
plt.plot(test, marker='.', label='cnt', color = 'black')
plt.plot(gru, marker='.', label='GRU', color = 'green')
plt.legend()
````

````python
plt.figure(figsize=(15,5))
plt.plot(test, marker='.', label='cnt', color = 'black')
plt.plot(rnn, marker='.', label='RNN', color = 'red')
plt.plot(lstm, marker='.', label='LSTM', color = 'blue')
plt.plot(gru, marker='.', label='GRU', color = 'green')
plt.legend()
````

````python
from sklearn.metrics import mean_squared_error
def RMSE(y_test, y_predict):
    return np.sqrt(mean_squared_error(y_test, y_predict))
````

````python
print("RNN RMSE:", RMSE(test, rnn))

print("LSTM RMSE:", RMSE(test, lstm))

print("GRU RMSE:", RMSE(test, gru))
````

1. `lstm.compile(loss='mse', optimizer='adam', metrics=['mse'])`
 : Keras의 LSTM 모델을 컴파일하는 부분입니다. 여기서 `loss='mse'`는 모델의 손실 함수로 평균 제곱 오차(Mean Squared Error, MSE)를 사용하겠다는 설정을 나타냅니다. 모델이 훈련되는 동안 이 손실 함수를 최소화하도록 최적화되게 됩니다.

2. 사용자 정의 함수 `RMSE`
 : 이 함수는 두 개의 인자인 `y_test`와 `y_predict`를 받아서 RMSE(평균 제곱근 오차)를 계산하여 반환합니다. RMSE는 모델이 예측한 값과 실제 값 사이의 차이를 나타내는 지표로, 값이 작을수록 모델의 예측이 실제 값과 가깝다는 것을 나타냅니다. `mean_squared_error` 함수는 sklearn.metrics 모듈에서 제공되며, 실제 값과 예측 값 간의 평균 제곱 오차를 계산하는 데 사용됩니다.

두 코드의 차이점은 다음과 같습니다:
- 첫 번째 코드는 모델 훈련 중에 사용되는 손실 함수를 설정하는 역할을 합니다.
- 두 번째 코드는 모델의 예측 결과와 실제 값 간의 차이를 계산하는 사용자 정의 함수를 정의합니다.

만약 두 코드를 비교하면, 첫 번째 코드는 모델을 훈련하거나 컴파일할 때 사용되며, 두 번째 코드는 모델의 평가 결과를 평가 지표로 변환하기 위해 사용될 수 있습니다.
