# 1. 네이버 스포츠 뉴스

```python
import requests   # 웹요청 위한 requests 라이브러리
```

```python
nv_sports_url = "https://sports.news.naver.com/index.nhn"  # 크롤링대상 변수
nhn_sports_main = requests.get(nv_sports_url)  # requests.get() url 요청
```

```python
from bs4 import BeautifulSoup  # bs4의 라이브러리의 BeatifulSoup 함수를 import, bs4.BeautifulSoup() 과 동일
nhn_sports_main_list = BeautifulSoup(nhn_sports_main.content, "html.parser", from_encoding = "utf-8")
```

```python
print(nhn_sports_main.text)
```

1. 요소 태그 선택자: `tagname`
   - 예시: `nhn_sports_main_list.select('div')`
   - 페이지의 모든 `<div>` 요소를 선택합니다.

2. 클래스 선택자: `.class`
   - 예시: `nhn_sports_main_list.select('.title')`
   - `class` 속성 값이 "title"인 모든 요소를 선택합니다.

3. 아이디 선택자: `#id`
   - 예시: `nhn_sports_main_list.select('#content')`
   - `id` 속성 값이 "content"인 요소를 선택합니다.

4. 자식 선택자: `parent > child`
   - 예시: `nhn_sports_main_list.select('div > p')`
   - `<div>` 요소의 직계 자식으로 있는 모든 `<p>` 요소를 선택합니다.

5. 속성 선택자: `[attribute=value]`
   - 예시: `nhn_sports_main_list.select('a[href="https://example.com"]')`
   - `href` 속성 값이 "https://example.com"인 모든 `<a>` 요소를 선택합니다.

```python
link_list = nhn_sports_main_list.select('#content')  # 태그 선택자 활용 select
link_list
```

```python
print(link_list[0].text)
```

# 2. 디씨인사이드

```python
import requests
from bs4 import BeautifulSoup
dc = requests.get("https://www.dcinside.com/")
```

```python
html = dc.text
print(html)
```

```python
dc_main_list = BeautifulSoup(dc.content, "html.parser", from_encoding = "utf=8")
```

```python
link_list = dc_main_list.select('ul.typet_list > li')  # 태그 선택자 활용 select
print(link_list[0].text)
```

```python
print(link_list[0].select_one('p').text)
      # 링크 텍스트와 주소만 필요하다면 select_one 으로 접근 (차이점은 select는 결과값이 리스트 형태로 저장되고, select_one은 그렇지 않다.)
```
